# 02 — Perturbation generator (Stage 8)

Apply each error type to a sample report and compare outputs.


```python
from neuroval3d.evaluation.perturbation import PerturbationOp, perturb
import numpy as np
rng = np.random.default_rng(0)
report = ('There is a 3.5 cm avidly enhancing left frontal mass with marked oedema. '
          'Two satellite lesions are noted. No restricted diffusion. T1 imaging shows the lesion clearly.')
for op in PerturbationOp:
    rec = perturb(report, op, rng)
    print(op.value, '->', (rec.perturbed if rec else '<no candidate>'))
```
