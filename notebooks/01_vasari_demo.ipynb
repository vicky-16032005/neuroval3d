# 01 — VASARI lexicon + parser demo

Shows the 30-feature VASARI lexicon and parses a sample report into a feature vector.


```python
from neuroval3d.grounding import VASARI_FEATURES, VASARIParser, vasari_vocabulary
print(len(VASARI_FEATURES), 'features')
for f in VASARI_FEATURES[:6]:
    print(f.code, f.name, '->', f.values)
```

```python
vocab = __import__('neuroval3d.grounding.vasari', fromlist=['vasari_vocabulary']).vasari_vocabulary()
print(len(vocab), 'vocab tokens')
print(vocab[:30])
```

```python
report = '''There is a 3.5 cm avidly enhancing intra-axial mass in the left frontal lobe.
Marked surrounding oedema. No restricted diffusion.'''
feats = VASARIParser().parse(report).to_dict()
feats
```
