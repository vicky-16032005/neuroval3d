# 03 — Validator scores on a clean vs perturbed report

Demonstrates that the validator gives lower scores to hallucinated reports.


```python
from neuroval3d.validators import LexicalValidator, SemanticValidator, StructuralValidator, FusionValidator
```

```python
ref = 'Right frontal high-grade glioma, 3.5 cm, with moderate edema. No restricted diffusion.'
clean = ref
flipped = ref.replace('Right', 'Left').replace('No restricted diffusion', 'Restricted diffusion is present')

sem = SemanticValidator()
lex = LexicalValidator().fit([ref, flipped, clean])
st = StructuralValidator()

print('clean :', sem.score(ref, clean), lex.score(ref, clean), st.score(ref, clean))
print('flip  :', sem.score(ref, flipped), lex.score(ref, flipped), st.score(ref, flipped))
```
