# 03 — Seven-axis validator on clean vs perturbed reports

Demonstrates each axis independently and the fused score.


```python
from neuroval3d.validators import (
    SemanticValidator, LexicalValidator, StructuralValidator,
    NumericValidator, ModalityValidator, NegationValidator, LesionTypeValidator,
    FusionValidator,
)
```

```python
ref = 'Right frontal high-grade glioma, 3.5 cm, with marked oedema on FLAIR. No restricted diffusion.'
clean = ref
flipped = ref.replace('Right', 'Left').replace('No restricted diffusion', 'Restricted diffusion is present').replace('glioma', 'meningioma').replace('3.5 cm', '1.0 cm').replace('FLAIR', 'T1')

sem = SemanticValidator()
lex = LexicalValidator().fit([ref, flipped, clean])
st = StructuralValidator()
num = NumericValidator()
mod = ModalityValidator()
neg = NegationValidator()
lt = LesionTypeValidator()

for label, cand in [('clean', clean), ('flipped', flipped)]:
    print(f'\n--- {label} ---')
    print('semantic   :', round(sem.score(ref, cand), 3))
    print('lexical    :', round(lex.score(ref, cand), 3))
    print('structural :', round(st.score(ref, cand), 3))
    print('numeric    :', round(num.score(ref, cand), 3))
    print('modality   :', round(mod.score(ref, cand), 3))
    print('negation   :', round(neg.score(ref, cand), 3))
    print('lesion_type:', round(lt.score(ref, cand), 3))
```
