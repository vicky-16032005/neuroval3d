# 04 — Perturbation benchmark + AUROC table

Headline notebook for the paper. Produces the AUROC table over the 7 NeuroVal-3D axes plus baselines on a perturbation set of 80 synthetic reports × 4 perturbed variants each = 480 records.

Expected (n=80, CPU, with BioClinicalBERT load): fusion ≈ 0.88, lexical ≈ 0.61, structural ≈ 0.67, semantic ≈ 0.25, RaTEScore-lite ≈ 0.06.


```python
from neuroval3d.evaluation import run_benchmark
result = run_benchmark(use_synthetic=True, n_samples=80, n_per_report=4)
print(result.summary_table())
```

```python
import json
print('=== overall AUROC ===')
print(json.dumps(result.auroc_overall, indent=2))
print('\n=== AUROC by perturbation op ===')
print(json.dumps(result.auroc_by_op, indent=2))
```
